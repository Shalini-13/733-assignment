```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('/content/student_data.csv')

# Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='Hours_Studied', y='Results', hue='Review_Session')
plt.title('Scatter Plot of Hours Studied vs Results')
plt.show()

# Fit model (using the entire dataset for simplicity)
X = data[['Hours_Studied', 'Review_Session']]
y = data['Results']

model = LogisticRegression()
model.fit(X, y)

# Output model coefficients
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)

# Predict on the same dataset
predictions = model.predict(X)

# Performance metrics
accuracy = accuracy_score(y, predictions)
auc = roc_auc_score(y, model.predict_proba(X)[:,1])

print(f"Accuracy: {accuracy}")
print(f"AUC-ROC: {auc}")
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming you want to use the same dataset for SVM
X = data[['Hours_Studied', 'Review_Session']]
y = data['Results']

# Linear SVM
linear_svm = svm.SVC(kernel='linear')
linear_svm.fit(X, y)

# RBF SVM with Grid Search
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 10]}
rbf_svm = svm.SVC(kernel='rbf')
grid_search = GridSearchCV(rbf_svm, param_grid, cv=5)
grid_search.fit(X, y)

print("Best Parameters for RBF SVM: ", grid_search.best_params_)
print("Best Score for RBF SVM: ", grid_search.best_score_)
```
